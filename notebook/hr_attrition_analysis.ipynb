```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
warnings_import = __import__('warnings')
warnings_import.filterwarnings('ignore')
```

```python
# Load data
df = pd.read_csv('WA_Fn-UseC_-HR-Employee-Attrition.csv')
```

```python
print("Shape:", df.shape)
print("\nAttrition counts:")
print(df['Attrition'].value_counts())
print("\nAttrition rate:", round(df['Attrition'].value_counts(normalize=True)['Yes']*100, 2), "%")
print("\nColumns:", df.columns.tolist())
```

```python
df['Attrition_Binary'] = (df['Attrition'] == 'Yes').astype(int)
```

```python
%matplotlib inline
import matplotlib.pyplot as plt

df['Attrition_Binary'] = (df['Attrition'] == 'Yes').astype(int)

fig, axes = plt.subplots(2, 3, figsize=(18, 10))
fig.suptitle('HR Attrition Analysis', fontsize=16, fontweight='bold')

# 1. Department
dept = df.groupby('Department')['Attrition_Binary'].mean() * 100
axes[0,0].bar(dept.index, dept.values, color=['#E74C3C','#3498DB','#2ECC71'])
axes[0,0].set_title('Attrition Rate by Department')
axes[0,0].set_ylabel('Attrition Rate (%)')

# 2. Overtime
ot = df.groupby('OverTime')['Attrition_Binary'].mean() * 100
axes[0,1].bar(ot.index, ot.values, color=['#2ECC71','#E74C3C'])
axes[0,1].set_title('Attrition Rate by Overtime')

# 3. Marital Status
marital = df.groupby('MaritalStatus')['Attrition_Binary'].mean() * 100
axes[0,2].bar(marital.index, marital.values, color=['#E74C3C','#3498DB','#2ECC71'])
axes[0,2].set_title('Attrition Rate by Marital Status')

# 4. Age
axes[1,0].hist(df[df['Attrition']=='Yes']['Age'], bins=20, alpha=0.7, color='#E74C3C', label='Left')
axes[1,0].hist(df[df['Attrition']=='No']['Age'], bins=20, alpha=0.7, color='#3498DB', label='Stayed')
axes[1,0].set_title('Age Distribution')
axes[1,0].legend()

# 5. Income
axes[1,1].boxplot([df[df['Attrition']=='Yes']['MonthlyIncome'],
                   df[df['Attrition']=='No']['MonthlyIncome']],
                   labels=['Yes','No'])
axes[1,1].set_title('Monthly Income by Attrition')

# 6. Job Role
role = df.groupby('JobRole')['Attrition_Binary'].mean() * 100
role = role.sort_values(ascending=True)
axes[1,2].barh(role.index, role.values, color='#E74C3C')
axes[1,2].set_title('Attrition by Job Role')

plt.tight_layout()
plt.savefig('hr_attrition_charts.png', dpi=150, bbox_inches='tight')
plt.show()
```

```python
%matplotlib inline
import matplotlib.pyplot as plt
import seaborn as sns

# Select numeric columns only
numeric_cols = ['Age', 'DailyRate', 'DistanceFromHome', 'Education',
                'EnvironmentSatisfaction', 'HourlyRate', 'JobInvolvement',
                'JobLevel', 'JobSatisfaction', 'MonthlyIncome',
                'NumCompaniesWorked', 'PercentSalaryHike',
                'StockOptionLevel', 'TotalWorkingYears',
                'TrainingTimesLastYear', 'WorkLifeBalance',
                'YearsAtCompany', 'YearsInCurrentRole',
                'YearsSinceLastPromotion', 'YearsWithCurrManager',
                'Attrition_Binary']

corr = df[numeric_cols].corr()['Attrition_Binary'].sort_values()

plt.figure(figsize=(10, 8))
corr.drop('Attrition_Binary').plot(kind='barh',
          color=['#E74C3C' if x < 0 else '#3498DB' for x in corr.drop('Attrition_Binary')])
plt.title('Correlation with Attrition', fontsize=14, fontweight='bold')
plt.xlabel('Correlation Coefficient')
plt.axvline(x=0, color='black', linewidth=0.8)
plt.tight_layout()
plt.savefig('attrition_correlation.png', dpi=150, bbox_inches='tight')
plt.show()
print("Saved.")
```
